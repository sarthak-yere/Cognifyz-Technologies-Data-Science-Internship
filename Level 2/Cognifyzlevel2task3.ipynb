Task: Feature Engineering
Extract additional features from the existing columns, such as the length of the restaurant name or address.
Create new features like "Has Table Booking"or "Has Online Delivery" by encodingcategorical variables.

1. Extracting Additional Features

```python
import pandas as pd
df=pd.read_csv('Dataset .csv')
```

a. Length of Restaurant Name

```python
df['Restaurant Name Length']=df['Restaurant Name'].apply(len)
```

b. Length of Address

```python
df['Address Length']=df['Address'].apply(len)
```

c. Number of Cuisines

```python
df['Cuisines']=df['Cuisines'].fillna('').astype(str)
df['Number of Cuisines'] = df['Cuisines'].apply(lambda x: len(x.split(', '))if x else 0)
df
```

2. Encoding Categorical Variables

a. Has Table Booking Encoding

```python
df['Has Table booking']=df['Has Table booking'].map({'Yes':1,'No':0})
```

b. Has Online Delivery Encoding

```python
df['Has Online delivery']=df['Has Online delivery'].map({'Yes':1,'No':0})
```

c. Price Range Encoding

```python
df['Price range']=df['Price range'].astype('category').cat.codes
print(df[['Has Table booking', 'Has Online delivery', 'Price range']].head(20))
```
