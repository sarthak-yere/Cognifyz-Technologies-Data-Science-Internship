Task: Price Range Analysis
Determine the most common price range
among all the restaurants.
Calculate the average rating for each price
range.
Identify the color that represents the highest
average rating among different price ranges.

1. Determine the Most Common Price Range among all the restaurants.

```python
import pandas as pd
df=pd.read_csv('Dataset .csv')
most_common_price_range=df['Price range'].mode()[0]
print(f"Most common Price Range among all the restaurants:{most_common_price_range}")
```

2.Calculate the average rating for each price
range.

```python
import matplotlib.pyplot as plt
import seaborn as sns
average_rating_for_each_price_range=df.groupby('Price range')['Aggregate rating'].mean().sort_values(ascending=False)
print("Average rating for each price range:")
print(average_rating_for_each_price_range)
plt.figure(figsize=(10,6))
sns.barplot(x=average_rating_for_each_price_range.index,y=average_rating_for_each_price_range.values)
plt.xlabel("Average Ratings")
plt.ylabel("Price Range")
plt.title(" Average rating for each price range")
plt.show()
```

3.Identify the color that represents the highest average rating among different price ranges.

```python
avg_rating_color_by_price_range=df.groupby('Price range').agg(
    average_rating=('Aggregate rating','mean'),
    rating_color=('Rating color',lambda x:x.mode()[0])#finding most common color for each price range
).sort_values(by='average_rating',ascending=False)
print("The color that represents the highest average rating among different price ranges:")
print(avg_rating_color_by_price_range)
    
```
