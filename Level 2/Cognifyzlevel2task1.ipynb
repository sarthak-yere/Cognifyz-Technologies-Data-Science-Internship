Task: Table Booking and Online Delivery
Determine the percentage of restaurants that offer table booking and online delivery.
Compare the average ratings of restaurantswith table booking and those without.
Analyze the availability of online delivery among restaurants with different price ranges.

1. Determining the percentage of restaurants offering table booking and online delivery

i) Percentage of restaurants with table booking and Online Delivery

```python
import pandas as pd
df=pd.read_csv('Dataset .csv')
total_restaurants=len(df)
print('Total Restaurants:',total_restaurants)
#for Table Booking
restaurants_with_table_booking=df[df['Has Table booking']=='Yes'].shape[0]

percentage_table_booking=(restaurants_with_table_booking/total_restaurants)* 100

print("The percentage of restaurants having table booking:",percentage_table_booking,'%')

#for Online Delivery

restaurants_with_online_delivery=df[df['Has Online delivery']=='Yes'].shape[0]
percentage_online_delivery=(restaurants_with_online_delivery/total_restaurants)* 100
print("The percentage of restaurants having online Delivery:",percentage_online_delivery,'%')
```

```python
import matplotlib.pyplot as plt
#for Table Booking
table_booking_counts=df['Has Table booking'].value_counts()
plt.figure(figsize=(7,7))
plt.pie(table_booking_counts, labels=table_booking_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99'])
plt.title('Table Booking availability')
plt.show()

#for Table Booking
online_delivery_counts =df['Has Online delivery'].value_counts()
plt.figure(figsize=(7,7))
plt.pie(online_delivery_counts, labels=online_delivery_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff6666', '#ffcc99'])
plt.title('Online Booking Availability')
plt.show()
```

2. Comparing the average ratings of restaurants with table booking vs. those without

```python
avg_rating_for_table_booking_availability=df[df['Has Table booking']=='Yes']['Aggregate rating'].mean()
avg_rating_for_table_booking_unavailability=df[df['Has Table booking']=='No']['Aggregate rating'].mean()
print(f'Average Rating with table Booking:{avg_rating_for_table_booking_availability}')
print(f'Average Rating with table Booking:{avg_rating_for_table_booking_unavailability}')

#Visualization(Bar Plot)

avg_ratings=[avg_rating_for_table_booking_availability,avg_rating_for_table_booking_unavailability]
labels=['With Table Booking','Without Table Booking']
plt.figure(figsize=(10,6))
plt.bar(labels, avg_ratings, color=['#ffcc66', '#66ccff'])
plt.title("Average ratings of restaurants with table booking vs. those without")
plt.ylabel('Average Ratings')
plt.show()
```

3. Analyzing the availability of online delivery among restaurants with different price ranges

```python
online_delivery_by_price_range=df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack().fillna(0)
print('Online Delivery among restaurants with different price ranges',online_delivery_by_price_range)
online_delivery_by_price_range.plot(kind='bar',stacked=True,color=['#ff6666', '#66b3ff'],figsize=(10,6))
plt.title('Availability of Online delivery by price range')
plt.xlabel('Price Range')
plt.ylabel('Proportion of Restaurants')
plt.legend(title="Online Delivery",labels=['No','Yes'])
plt.show()
```
