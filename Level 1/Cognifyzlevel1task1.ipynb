:- Checking for the missing values

```python
#Total number of missing values in a dataset
missing_values=df.isnull().sum()
print(missing_values)
```

*Here the cuisine column contains 9 missing values thus we can replace the missing values with the most frequent cuisine based on the other values in the column.This can be done by using mode.

```python
mode_value=df['Cuisines'].mode()[0]#get the most frequent cuisine
print(mode_value)
```

```python
df['Cuisines']=df['Cuisines'].fillna(mode_value)
```

*Dataset after replacing missing values

```python
missing_values=df.isnull().sum()
print(missing_values)
```

Since the Aggregate rating column is already of the correct float datatype, we can proceed directly with analyzing the distribution and checking for any class imbalances.

1. Analyze the Distribution of the Aggregate Rating:

```python
import matplotlib.pyplot as plt
import seaborn as sns
#creating a histogram for aggregate rating

plt.figure(figsize=(10,6))
sns.histplot(df['Aggregate rating'],kde=True,bins=30)
plt.title("Aggregate Rating Distribution")
plt.xlabel("Aggregate Ratings")
plt.ylabel("Frequency")
plt.show()
```

2. Check for Class Imbalances.(It is'nt required while dealing with continuous data)(Optional)

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Aggregate rating',data=df)
plt.title('Class Distribution for Aggregate Ratings')
plt.xlabel('Aggregate Ratings')
plt.ylabel('count')
plt.show()
```

3.Here exits a class imbalance as most of the ratings are concentrated near '0'.
